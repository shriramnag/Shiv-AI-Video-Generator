```
# @title 🚀 १. सेटअप: गिटहब से कोड लोड करें
import os
import shutil

REPO_URL = "https://github.com/shriramnag/Shiv-AI-Video-Generator"
REPO_NAME = "Shiv-AI-Video-Generator"

if os.path.exists(REPO_NAME):
    shutil.rmtree(REPO_NAME)

print("गिटहब से कोड लोड हो रहा है...")
!git clone {REPO_URL}
%cd {REPO_NAME}

print("\nजरूरी लाइब्रेरी इंस्टॉल हो रही हैं...")
!pip install -q -r requirements.txt
!pip install -q diffusers transformers accelerate huggingface_hub gradio
print("\n✅ सेटअप पूरा हुआ!")
```

```
# @title 🔑 २. मॉडल डाउनलोड (No Token Required)
from huggingface_hub import snapshot_download

MODEL_ID = "Shriramnag/Shiv-AI-Video-Generator"
print(f"\nहगिंग फेस से {MODEL_ID} पब्लिकली डाउनलोड हो रहा है...")

try:
    # बिना टोकन के डाउनलोड करने का लॉजिक
    snapshot_download(
        repo_id=MODEL_ID,
        local_dir="./shiv_model",
        use_auth_token=False,  # टोकन की ज़रूरत को खत्म किया
        ignore_patterns=["*.msgpack", "*.bin", "*.h5"]
    )
    print("\n✅ मॉडल सफलतापूर्वक डाउनलोड हो गया!")
except Exception as e:
    print(f"\n❌ एरर: सुनिश्चित करें कि हगिंग फेस पर आपका मॉडल 'Public' सेट है।")
    print(f"Error details: {e}")
```
